# Y! Watchlist



```python
import requests
import urllib
from bs4 import BeautifulSoup
from collections import Counter

baseUrl = "https://finance.yahoo.com/u/yahoo-finance/watchlists/most-active-penny-stocks"
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:101.0) Gecko/20100101 Firefox/101.0', "cookie": "CONSENT=YES+cb.20230531-04-p0.en+FX+908"}

response = requests.get(baseUrl , headers=headers )

if response.status_code != 200:
	print("Error fetching page")
	exit()
else:

	content = response.content
print(content)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# The title tag of the page
print(soup.title)
```

```python
# All links in the page
nb_links = len(soup.find_all('a'))
print(f"There are {nb_links} links in this page")
```

```python
# Text from the page
print(soup.get_text())
```

```python
first_link = soup.a
print(first_link)
```

```python
    # MOST COMMON LINKS

all_hrefs = [a.get('href') for a in soup.find_all('a')]
top_3_links = Counter(all_hrefs).most_common(3)
print(top_3_links)
```

```python
# ALL LINKS

all_hrefs = [a.get('href') for a in soup.find_all('a')]
allLinks = Counter(all_hrefs)
print('All Links:', allLinks)
```

```python
# The text of the link


print(first_link.text)


# The href of the link
print(first_link.get('href'))
```

```python
#   cwl-symbols W(100%)

cwlSymbols = soup.find(class_='cwl-symbols')
print('CWL Symbols: \n',  cwlSymbols)

```

```python
#   data-col0 Ta(start) Pstart(6px) Pend(6px)

col0Symbols = soup.find(class_='data-col0')
print('Data Col 0: \n' , col0Symbols)
```

```python
#   data-col0 Ta(start) Pstart(6px) Pend(6px)

allCol0Symbols = soup.find_all(class_='data-col0')
print('Data Col 0: \n' , allCol0Symbols)
```
